```python
%%writefile student_management.py

students = []
#add student data
def add_student_bio(name, age, grade):
     student = {"name": name, "age": age, "grade": grade}
     students.append(student)
     return student

#show all student data
def all_students_bio():
  return students

#remove student data
def remove_student_bio(name):
    for student in students:
        if student["name"] == name:
            students.remove(student)
            return True
    return False

#get individual student data by name
def get_student_bio_by_name(name):
    for student in students:
        if student["name"] == name:
            return student
    return None

students_marks = []

# Add student marks and get their total,percentage,grade
def add_student_marks_by_name(name,maths,science,english):
    total_marks = maths + science + english
    percentage = total_marks / 300 * 100
    if percentage >= 90:
        grade = "A"
    elif percentage >= 80:
        grade = "B"
    elif percentage >= 70:
        grade = "C"
    student_marks = {"name": name, "maths": maths, "science": science,"english":english}
    student_marks["total_marks"] = total_marks
    student_marks["percentage"] = percentage
    student_marks["grade"] = grade
    students_marks.append(student_marks)
    return student_marks

# all student marks,total,percentage,grade
def all_student_marks():
    return students_marks

#remove student marks
def remove_student_marks(name):
    for student_marks in students_marks:
        if student_marks["name"] == name:
            students_marks.remove(student_marks)
            return True
    return False

# topper of class
def class_topper():
    topper = None
    highest_percentage = 0
    for student_marks in students_marks:
        if student_marks["percentage"] > highest_percentage:
            highest_percentage = student_marks["percentage"]
            topper = student_marks
    return topper

# average of class
def class_average():
    total_marks = 0
    for student_marks in students_marks:
        total_marks += student_marks["total_marks"]
        average_marks = total_marks / len(students_marks)
    return average_marks

# update student marks
def update_student_marks(name, maths, science, english):
    for student_marks in students_marks:
        if student_marks["name"] == name:
            student_marks["maths"] = maths
            student_marks["science"] = science
            student_marks["english"] = english
            student_marks["total_marks"] = maths + science + english
            student_marks["percentage"] = (maths + science + english) / 300 * 100
            if student_marks["percentage"] >= 90:
                student_marks["grade"] = "A"
            elif student_marks["percentage"] >= 80:
                student_marks["grade"] = "B"
            elif student_marks["percentage"] >= 70:
                student_marks["grade"] = "C"
            return student_marks
    return None


```
