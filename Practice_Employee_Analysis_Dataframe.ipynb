```python
import pandas as pd
import numpy as np
```

```python
data ={
      "Employee_id" : ["EMP001","EMP002","EMP003","EMP004","EMP005"],
      "Name" : ["Narendra","Rahul","Priya","Amit","Neha"],
      "Department" : ["IT","HR","Finance","IT","HR"],
      "Days_present" : [24,20,26,18,22],
      "Days_absent" : [2,6,0,8,4],
      "Performance_Score":[88,75,95,65,82]
      }

df = pd.DataFrame(data)
df
```

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull())
```

```python
#Attendance analysis

df["Attendance_pct"] = (df['Days_present']/(df['Days_present']+df['Days_absent'])*100).round(2)
df
```

```python
#performance analysis

def performance_analysis(performance_score):
    if performance_score >= 90:
      return 'Excellent'
    elif performance_score >= 80:
      return 'Good'
    elif performance_score >= 70:
      return 'Average'
    elif performance_score < 70:
      return 'Needs improvment'

df['Performance_Category'] = df['Performance_Score'].apply(performance_analysis)
df
```

```python
#department analysis

#1.Avg performance_score by department

def department_avg_score(department):
   avg_score = df[df['Department'] == department]['Performance_Score'].mean()
   return avg_score
print(department_avg_score('IT'))

#2.Avg attendance by department
def department_avg_attendance(department):
    avg_attendance =df[df['Department'] == department]['Attendance_pct'].mean().round(2)
    return avg_attendance
print(department_avg_attendance('IT'))

#3.Employee count
def employee_count(department):
    count = df[df['Department'] == department]['Employee_id'].count()
    return count
print(employee_count('IT'))
```

```python
#1. top 3 employees by performance_score
df.sort_values(by='Performance_Score',ascending=False).head(3)

#2. employees with attendance less 80%
print(df[df['Attendance_pct'] < 80])

df.groupby('Department')['Attendance_pct'].mean().round(2)
# df.loc[(df['Attendance_pct']<80),['Name','Attendance_pct']]

# employees with performance above department performance avg
df['dept_avg'] = df.groupby("Department")["Performance_Score"].transform('mean').round(2)

df

# students above average

df[df['Performance_Score'] > df['dept_avg']]

```

```python
performance_score = np.array(df['Performance_Score'])

performance_score.mean()
performance_score.max()
performance_score.min()
performance_score.std()
```

```python
# dashboard
print("--DashBoard--")
count = df['Employee_id'].count()
print(f"Total employees: {count}")
Highest_performance = df['Performance_Score'].max()
print(f"Highest performance: {Highest_performance}")
Lowest_performance = df['Performance_Score'].min()
print(f"Lowest performance: {Lowest_performance}")
avg_attendance = df['Attendance_pct'].mean().round(2)
print(f"Average attendance: {avg_attendance}")
best_department = df.groupby('Department')['Performance_Score'].mean().idxmax()
print(f"Best department: {best_department}")
Need_imp_count = df[df['Performance_Category']=='Needs improvment']['Name'].count()
print(f"Number of employees need improvement: {Need_imp_count}")
```
