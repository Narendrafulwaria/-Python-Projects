**Problem Statement: Retail Store Product Management and Revenue Analytics System**


A retail store owner wants to build a Python-based inventory and sales analytics system to track products sold across different categories and generate business insights from sales data.

The system should allow users to add products with details such as product name, category, selling price, and quantity sold. For every product entered, the system must automatically calculate the revenue generated and store the information in a centralized product database.

The application should provide the following capabilities:

**Product Management**
Add new products along with category, price, and quantity sold.
Calculate and store revenue for each product.
Display all available product records.
Search and generate a report for a specific product by name.
**Revenue Analytics**
Calculate the total revenue generated by all products.
Identify the product with the highest revenue.
Identify the product with the lowest revenue.
Calculate the average revenue per product.
Calculate the total quantity of products sold.
**Category Analysis**
Display all products belonging to a specific category.
Calculate total revenue generated by a selected category.
Calculate total quantity sold within a selected category.
**Revenue Classification**
Classify products into revenue segments based on the revenue generated:
High Revenue: Revenue greater than or equal to ₹100,000
Medium Revenue: Revenue between ₹50,000 and ₹99,999
Low Revenue: Revenue below ₹50,000
